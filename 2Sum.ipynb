```python
def twoSumNaive(targetL,numb):
    #On^2 implementation
    for i in targetL:
        rest=int(numb-i)
        #the following booleans are type safe and always return a boolean
        if rest in intDict and rest!=i:
            return(i,rest)
        
    return None

def twoSumHash(targetL,numb):
    #Hashing implementation
    #Look up is constant time for 1 lement, so linear in the list
    intDict={}
    #build hashtable
    for i in targetL:
        intDict[i]=True
        
    for i in targetL:
        rest=int(numb-i)
        #the following booleans are type safe and always return a boolean
        if rest in intDict and rest!=i:
            return(i,rest)
        
    return None 
```

```python
#Let's try not having to rebuild the dictionary each time
#This version works much better
def twoSumHashG(numb):
    #Hashing implementation
    #Look up is constant time for 1 lement, so linear in the list
    global targetDict
    #build hashtable
        
    for i,j in targetDict.items():
        rest=int(numb-i)
        #the following booleans are type safe and always return a boolean
        if rest in targetDict and rest!=i:
            return(i,rest)
        
    return None 
```

```python
# our two sum function which will return
# all pairs in the list that sum up to S
def twoSum(S):
    global arr
    sums = []
    hashTable = {}

  # check each element in array
    for i in range(0, len(arr)):

    # calculate S minus current element
        sumMinusElement = S - arr[i]

    # check if this number exists in hash table
    # if so then we found a pair of numbers that sum to S
        if sumMinusElement in hashTable:
            sums.append([arr[i], sumMinusElement])

    # add the current number to the hash table
        hashTable[arr[i]] = arr[i]

  # return all pairs of integers that sum to S
    return sums
```
